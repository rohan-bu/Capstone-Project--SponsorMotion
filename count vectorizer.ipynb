```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
import string
from nltk.tokenize import word_tokenize
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
import re
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
nltk.download('stopwords')

import nltk
nltk.download('punkt')
```

```python
# Preprocess the data
data['summary'] = data['summary'].astype(str)
data['start date'] = data['start date'].astype(str)
data['location'] = data['location'].astype(str)
data['name'] = data['name'].astype(str)
```

```python
# Convert the summary column to lowercase
data['summary'] = data['summary'].str.lower()

# Remove punctuation
data['summary'] = data['summary'].str.replace('[{}]'.format(string.punctuation), '')

# Tokenize the summary column
data['summary'] = data['summary'].apply(word_tokenize)

# Remove stopwords
stop_words = set(stopwords.words('english'))
data['summary'] = data['summary'].apply(lambda x: [word for word in x if word not in stop_words])

# Flatten the list of tokenized words
words = [word for sublist in data['summary'] for word in sublist]

# Create a frequency distribution of words
freq_dist = FreqDist(words)

# Get the most frequent words and their frequencies
most_common = freq_dist.most_common(30)

# Print the most frequent words and their frequencies
for word, frequency in most_common:
    print(f'{word}: {frequency}')
```

```python
# Exclude most frequent words from the summary column
data['summary'] = data['summary'].apply(lambda x: ' '.join([word for word in x if word not in most_common]))

# Convert the columns to lowercase strings
data['start date'] = data['start date'].astype(str).str.lower()
data['location'] = data['location'].astype(str).str.lower()
data['name'] = data['name'].astype(str).str.lower()

# Concatenate columns into a single text column
data['combined_text'] = data['summary'] + ' ' + data['start date'] + ' ' + data['location'] + ' ' + data['name']

# Convert the combined_text column to a list
documents = data['combined_text'].tolist()

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the documents
X = vectorizer.fit_transform(documents)

# Compute cosine similarity matrix
cosine_sim = cosine_similarity(X)

# Find duplicates based on similarity scores
duplicates = []
for i in range(len(cosine_sim)):
    for j in range(i+1, len(cosine_sim)):
        if cosine_sim[i, j] > 0.8:  # Adjust the threshold as needed
            duplicates.append((i, j))

# Print the duplicate pairs
for pair in duplicates:
    print(f"Duplicate pair: {pair}")
```
